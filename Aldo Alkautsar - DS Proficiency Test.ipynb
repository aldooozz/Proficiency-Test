```python
def segment_hour(x):
    segment = ''
    if  x['HOUR'] >= 0 and  x['HOUR'] <4 :
        segment = 'Late Night'
    elif x['HOUR'] >= 6 and x['HOUR'] < 10:
        segment = 'Early Morning'
    elif x['HOUR'] >= 10 and x['HOUR'] < 12:
        segment = 'Morning'
    elif x['HOUR'] >= 12 and x['HOUR'] < 14:
        segment = 'Ealy Afternoon '
    elif x['HOUR'] >= 14 and x['HOUR'] < 17:
        segment='Afternoon'
    elif x['HOUR'] >= 17 and x['HOUR'] < 20:
        segment='Evening' 
    elif x['HOUR'] >= 20 and x['HOUR'] < 24:
        segment='Night'
    else:
        segment = 'not_specified'
    return segment

df['Time'] = df.apply(lambda x : segment_hour(x), axis=1)
df.head(2)
```
